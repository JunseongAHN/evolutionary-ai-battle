# PR3: PPO Smoke Training

This section trains through the notebook, saves reward-selected checkpoints, and records `checkpoint_selected.pt`. Adjust the notebook training knobs in the setup cell before running this cell.

```python
import json
import shlex
import subprocess
import sys
from pathlib import Path

PROJECT_ROOT = PROJECT_ROOT if 'PROJECT_ROOT' in globals() else Path.cwd()
EXPERIMENT_ROOT = EXPERIMENT_ROOT if 'EXPERIMENT_ROOT' in globals() else PROJECT_ROOT / 'experiment'
if not (EXPERIMENT_ROOT / 'training').exists():
    candidates = [
        PROJECT_ROOT if PROJECT_ROOT.name == 'experiment' else PROJECT_ROOT / 'experiment',
        Path.cwd() if Path.cwd().name == 'experiment' else Path.cwd() / 'experiment',
        Path.cwd().parent / 'experiment',
    ]
    for candidate in candidates:
        if (candidate / 'training').exists():
            EXPERIMENT_ROOT = candidate.resolve()
            PROJECT_ROOT = EXPERIMENT_ROOT.parent
            break
    else:
        raise ModuleNotFoundError(f'Could not find experiment/training from cwd={Path.cwd()}')
CONFIG_PATH = EXPERIMENT_ROOT / 'configs' / 'local_combat_in_range.yaml'

def run_command(cmd):
    print('running:')
    print(' '.join(shlex.quote(str(part)) for part in cmd))
    process = subprocess.Popen(
        cmd,
        cwd=PROJECT_ROOT,
        text=True,
        stdout=subprocess.PIPE,
        stderr=subprocess.STDOUT,
        bufsize=1,
    )
    lines = []
    for line in process.stdout:
        print(line, end='')
        lines.append(line)
    return_code = process.wait()
    if return_code != 0:
        raise RuntimeError(f'train_ppo.py failed with exit code {return_code}')
    return ''.join(lines)

debug_samples = int(NOTEBOOK_DEBUG_RESET_SAMPLES)
if debug_samples > 0:
    debug_cmd = [
        sys.executable,
        str(EXPERIMENT_ROOT / 'train_ppo.py'),
        '--config',
        str(CONFIG_PATH),
        '--debug-reset-samples',
        str(debug_samples),
    ]
    debug_output = run_command(debug_cmd)
    print('reset debug complete; running training next.')

train_cmd = [
    sys.executable,
    str(EXPERIMENT_ROOT / 'train_ppo.py'),
    '--config',
    str(CONFIG_PATH),
    '--progress',
]
train_output = run_command(train_cmd)

output_text = train_output
final_json_start = output_text.rfind('\n{')
if final_json_start == -1:
    final_json_start = output_text.find('{')
else:
    final_json_start += 1
LAST_TRAIN_RESULT = json.loads(output_text[final_json_start:])
SELECTED_CHECKPOINT = LAST_TRAIN_RESULT.get('checkpoint_selected', LAST_TRAIN_RESULT.get('checkpoint_max_reward'))
LATEST_CHECKPOINT = LAST_TRAIN_RESULT.get('checkpoint_latest')

print('checkpoint_latest:', LATEST_CHECKPOINT)
print('checkpoint_selected:', SELECTED_CHECKPOINT)
print('checkpoint_max_reward:', LAST_TRAIN_RESULT.get('checkpoint_max_reward'))
```
